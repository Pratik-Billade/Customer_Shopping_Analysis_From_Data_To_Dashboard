## Data Exploration and Cleaning

```python
# importing the required libraries
import pandas as pd
```

```python
# loading the dataset into a pandas data frame
df = pd.read_csv('../dataset/raw/customer_shopping_analysis.csv')
```

```python
# viewing the first 5 rows of the dataset
df.head(5)
```

```python
# checking the count of rows and columns in the dataset
df.shape
```

```python
# checking the summary statistics of numeric columns
df.describe()
```

```python
# checking the summary statistics of all columns
df.describe(include='all')
```

```python
# chekcing the count of null values
df.isnull().sum()
```

```python
# filling missing 'Review Rating' values using the median rating of each 'Category'
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# verifying that missing values were filled
df.isnull().sum()
```

```python
# convert column names to lowercase snake_case
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

## Feature Engineering

```python
# creating a column age group
bins = [0, 25, 35, 50, 120]
labels = [
    "0-25 Young Adult",
    "26-35 Adult",
    "36-50 Middle-aged",
    "50+ Senior"
]

df["age_group"] = pd.cut(
    df["age"],
    bins=bins,
    labels=labels,
    right=True
)
```

```python
df[['age', 'age_group']]
```

```python
# counting how many records fall into each frequency_of_purchases group
df.groupby('frequency_of_purchases')['frequency_of_purchases'].count()
```

```python
# mapping purchase frequency labels to approximate number of days
frequency_mapping = {
    'Annually' : 365,
    'Bi-Weekly' : 14,
    'Every 3 Months' : 90,
    'Fortnightly' : 15,
    'Monthly' : 30,
    'Quarterly' : 90, 
    'Weekly' : 7   
}

# creating column purchase_frequency_in_days
df['purchase_frequency_in_days']  = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_in_days']].head(10)
```

```python
# checking for redundancy
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# removing the promo_code_used column from the data
df = df.drop('promo_code_used', axis = 1)
```

## Saving the cleaned data as a CSV file

```python
df.to_csv("../dataset/processed/customer_shopping_cleaned.csv", index = False)
```
